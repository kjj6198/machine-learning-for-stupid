```python
# import library
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
scores = [1.0, 2.0, 3.0]

def softmax(x):
    return np.exp(x) / np.sum(np.exp(x), axis = 0)

def meanized(x):
    return x / np.sum(x)

print(softmax(scores))
print(meanized(scores))
```

### Softmax
**softmax 跟一般標準化的差別在於他能夠將微小的變動放大。可以看到用 softmax 會讓整體的差異更加明顯；但一般的標準化就沒有這種效果了**

可以看到圖形。
放大輸出後，越接近 1 的那個機率，就會被明顯放大，而其他則會明顯減少。


```python
x = np.arange(-2.0, 6.0,0.1) # 產生等差數列，從 -2 到 6，等差為 0.1
scores = np.vstack([x, np.ones_like(x), 0.2 * np.ones_like(x)])

plt.plot(x, softmax(scores).T, linewidth = 2, label = "label")
plt.legend()
plt.show()
```
